```python
import cv2

video = cv2.VideoCapture('videos/2015_05_12_10_49_54FrontColor.avi')

while(video.isOpened()):
    ret, frame = video.read()
    if(ret == True):
        cv2.imshow("BUS", frame)
        if(cv2.waitKey(30) == ord('s')):
            break
    else:
        break
video.release()
cv2.destroyAllWindows()
```
